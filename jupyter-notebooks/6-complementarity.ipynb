### Study of complementarity between courses

The graph below represents the distribution of minor among students of a certain major.

```python
list_of_courses = list(df["CODE_COURS"].unique())

fig, axes = plt.subplots(5,2, facecolor="w", figsize=(30,30))
for course, ax in zip(list_of_courses, axes.flatten()):
    ax=fn.complementary_pie_chart(course, df, ax)
    ax.set_facecolor('w')
```
